##### 4. PyCBA
PyCBA is a python implementation of the Continuous Beam Analysis program, originally coded in Matlab and subsequently ported to C++ (by Pierrot).

PyCBA is for fast linear elastic analysis of general beam configuration

One of the main functions of PyCBA is that the basic analysis engine forms the basis for higher-level analysis. 


Current PyCBA includes modules for:

Influence line generation
Moving load analysis for bridges, targeted at bridge access assessments
Load patterning and envelopings.ns.

### 01. Example:

```python
# Define node locations, and support conditions
L = 25 # beam length in meters
# lets instalate or create a beam object using newEulerBeam2d method passing the length of the beam as an argument
beam = ps.newEulerBeam(L) # we are importing the plane Sections library
```

```python
# Define beam and fixities
pinned = [1,1,0]
beam.setFixity(0, pinned)
beam.setFixity(20, pinned)
```

```python
# Define point loads
Pz = -1000
beam.addLabel(0, label = "A")
beam.addLabel(10, label = "E")
beam.addVerticalLoad(15, -2*Pz, label = "D")
beam.addLabel(20, label = "B")
beam.addVerticalLoad(25, 3*Pz, label = "C")


# Define distibuted Loads
beam.addDistLoadVertical(0, L*0.4, 5*Pz)
```

```python
# Plot the beam diagram
ps.plotBeamDiagram(beam)

# Run the analysis
analysis = ps.OpenSeesAnalyzer2D(beam)
analysis.runAnalysis()
```

```python
# Plot the ShearForce and Bending Moment Diagramabs
ps.plotShear(beam, scale = 1, yunit = "kN")
ps.plotMoment(beam, scale = 1, yunit = "kNm")
```

### 06. Example:

```python
from indeterminatebeam import Beam
# Create 7 m beam with E, I, A as defaults
beam = Beam(7)                          
# Create 7 m beam with E, I, and A assigned by user
beam_2 = Beam(7, E=2000, I=10**6, A=3000) 
```

```python
from indeterminatebeam import Support
# Defines a pin support at location x = 5 m  
a = Support(0, (0,1,0))      
# Defines a roller support at location x = 0 m
b = Support(5, (1,1,0))      
# Defines a fixed support at location x = 7 m
c = Support(7, (1,1,1))      
# Assign the support objects to a beam object created earlier
beam.add_supports(a,b,c)    
```

```python
from indeterminatebeam import PointLoadV, PointTorque, DistributedLoadV
# Create a 1000 N point load at x = 2 m
load_1 = PointLoadV(1000, 2)
# Create a 2000 N/m UDL from x = 1 m to x = 4 m
load_2 = DistributedLoadV(2000, (1, 4))
# Defines a 2 kN.m point torque at x = 3.5 m
load_3 = PointTorque(2*10**3, 3.5)
# Assign the load objects to the beam object
beam.add_loads(load_1,load_2,load_3)
```

```python
beam.analyse()  
```

```python
fig_1 = beam.plot_beam_external()
fig_1.show()

fig_2 = beam.plot_beam_internal()
fig_2.show()
```
