### 1.5 What is Structural Analysis? 
Какво е структурен анализ?

Astructure is a system of interconnected members used to support external loads. Structural analysis is the prediction of the response of structures to specified arbitrary external loads. During the preliminary structural design stage, a structure’s potential external load is estimated, and the size of the structure’s interconnected members are determined based on the estimated loads. Structural analysis establishes the relationship between a structural member’s expected external load and the structure’s corresponding developed internal stresses and displacements that occur within the member when in service. This is necessary to ensure that the structural members satisfy the safety and the serviceability requirements of the local building code and specifications of the area where the structure is located.

Структурата е система от взаимосвързани елементи, използвани за поддържане на външни товари. Структурният анализ е прогнозирането на реакцията на конструкциите към определени произволни външни натоварвания. По време на етапа на предварителния структурен дизайн се оценява потенциалното външно натоварване на конструкцията и размерът на взаимосвързаните елементи на конструкцията се определя въз основа на прогнозните натоварвания. Структурният анализ установява връзката между очакваното външно натоварване на структурния елемент и съответните развити вътрешни напрежения и измествания на конструкцията, които възникват в елемента, когато е в експлоатация. Това е необходимо, за да се гарантира, че конструктивните елементи отговарят на изискванията за безопасност и експлоатационна годност на местния строителен кодекс и спецификациите на района, където се намира конструкцията.

### 1.6 What is Modal Analysis? 
Какво е модален анализ?

Modal analysis is an indispensable tool in understanding the structural dynamics of objects- how structures and objects vibrate and how resistant they are to applied forces. The modal analysis allows machines and structures to be tested, optimized, and validated. Natural resonance frequencies of the objects and damping parameters can be calculated, and mode shapes can be visualized on an animated geometry of the measured objects.

Модалният анализ е незаменим инструмент за разбиране на структурната динамика на обектите - как структурите и обектите вибрират и колко устойчиви са на приложените сили. Модалният анализ позволява машините и структурите да бъдат тествани, оптимизирани и валидирани. Естествените резонансни честоти на обектите и параметрите на затихване могат да бъдат изчислени, а формите на режима могат да бъдат визуализирани върху анимирана геометрия на измерените обекти.

### 1.7 Introduction to Finite Element Analysis 
Въведение в анализа на крайните елементи

Finite Element Analysis (FEA) is method used for the evaluation of structures, providing an accurate prediction of components response subjected to various structural loads. It is a computerized method for predicting how a product reacts to real world forces, vibration, heat, fluid flow and other physical effects. FEA works by breaking down a real object into a mesh of large number of finite elements, that combine to create shape of structure that is being assessed. Each of these small elements are subjected to calculations, with these mesh refinements combining to produce the final result of the whole structure.

Анализът на крайните елементи (FEA) е метод, използван за оценка на конструкции, осигуряващ точна прогноза за реакцията на компонентите, подложени на различни структурни натоварвания. Това е компютъризиран метод за прогнозиране как даден продукт реагира на реални сили, вибрации, топлина, поток от течности и други физически ефекти. FEA работи, като разделя реален обект на мрежа от голям брой крайни елементи, които се комбинират, за да създадат форма на структура, която се оценява. Всеки от тези малки елементи се подлага на изчисления, като тези уточнения на мрежата се комбинират, за да се получи крайният резултат от цялата структура.

### 1.8 Introduction to Beams: 

Beam is a horizontal structure member used to carry vertical load, shear load and sometime horizontal load. It is the major component of building structures. It mainly use in construction of bridges, trusses, and other structures which carry vertical load.

Гредата е хоризонтален структурен елемент, използван за носене на вертикално натоварване, натоварване на срязване и понякога хоризонтално натоварване. Това е основният компонент на строителните конструкции. Използва се главно при изграждането на мостове, ферми и други конструкции, които носят вертикално натоварване.

### Types of Beams in Constructions:

#### 1.8.1 Based on Support Conditions: 
Въз основа на условията за поддръжка:

##### 1. Simply Supported Beam: 
Просто греда:

Simply supported beam is supported at both end. One end of the beam is supported by hinge support and other one by roller support. This support allow to horizontal movement of beam. It beam type undergoes both shear stress and bending moment.

Просто греда се поддържа от двата края. Единият край на гредата се поддържа от шарнирна опора, а другият от ролкова опора. Тази опора позволява хоризонтално движение на гредата. Този тип греда претърпява както напрежение на срязване, така и момент на огъване

##### 2. Fixed Beam:
Запъната греда:

This beam is fixed from both ends. It does not allow vertical movement and rotation of the beam. It is only under shear stress and no moment produces in this beams.

Тази греда е фиксирана от двата края. Не позволява вертикално движение и въртене на гредата. То е само под напрежение на срязване и в тези греди не се създава момент.

##### 3. Cantilever Beam:
Конзолна греда:

Cantilever beams a structure member of which one end is fixed and other is free. This is one of the famous type of beam use in trusses, bridges and other structure member.

Конзолни греди Конструкционен елемент, на който единият край е фиксиран, а другият е свободен. Това е един от известните видове използване на греди в ферми, мостове и други конструктивни елементи.

##### 4. Continuous Beam:
Непрекъсната греда:

This beam is similar to simply supported beam except more than two support are used on it. One end of it is supported by hinged support and other one is roller support. One or more supports are use between these beams. It is used 16 in long concrete bridges where length of bridge is too large.

Тази греда е подобна на просто поддържаната греда, с изключение на това, че върху нея се използват повече от две опори. Единият му край се поддържа от шарнирна опора, а другият е ролкова опора. Една или повече опори се използват между тези греди. Използва се 16 при дълги бетонни мостове, където дължината на моста е твърде голяма.

##### 5. Over Hanging Beam:
Обратна греда:

Overhanging beam is combination of simply supported beam and cantilever beam. One or both of end overhang of this beam. This beam is supported by roller support between two ends.

Надвиснала греда е комбинация от просто поддържана греда и конзолна греда. Един или и двата крайни надвеса на тази греда. Тази греда се поддържа от ролкова опора между два края.

##### 6. Reinforced Concrete Beam:
Стоманобетонна греда:

It is constructed from concrete and reinforcement.Sometimes reinforced concrete beam is concealed in reinforced concrete slabs and it is called hidden beam or concealed beam.

Изгражда се от бетон и армировка. Понякога стоманобетонната греда е скрита в стоманобетонни плочи и се нарича скрита греда или скрита греда.

#### 1.8.4 Based on equilibrium condition:

Въз основа на условието за равновесие:

##### 1. Statically Determinate Beam: 
Статично определен греда:

For a statically determinate beam, equilibrium conditions alone can be used to solve reactions. The number of unknown reactions is equal to the number of equations.

##### 2. Statically Indeterminate Beam: 
Статично неопределена греда:

### 1.9 Types of load acting on beam:
Видове натоварване, действащо върху гредата:

Abeam is usually horizontal member and load which will be acting over the beam will be usually vertical loads. There are following types of loads:

And the sum of the moments about any point must be equal to zero. If we choose to take the moments about Point B we get the following:

И сумата от моментите около всяка точка трябва да е равна на нула. Ако изберем да вземем моментите за точка Б, получаваме следното:

$$ \sum M_B = 0 $$ 
$$ - ( R_A . x ) + F_1 ( x - 2 ) + F_2 ( x - 4 ) = 0 $$
$$ - ( R_A . 6 ) + 15.4 + 6.2 = 0 $$
$$  - ( R_A . 6 ) + 60 + 12 = 0 $$
$$ R_A = \frac{72}{6} $$
$$ R_A = 12 kN$$

We  take the moments about Point A we get the following:

Вземаме моментите за точка А и получаваме следното:

$$ \sum M_A = 0 $$ 
$$ - ( R_B . x ) + F_1 ( x - 4 ) + F_2 ( x - 2 ) = 0 $$
$$ - ( R_B . 6 ) + ( 15 . 2 ) + ( 6 . 4 ) = 0 $$
$$ - ( R_B . 6 ) = - 30 + 24 $$
$$ R_B = \frac{54}{6} $$
$$ R_B = 9 kN $$

!!!!!!!!!!!!!|All of the external loads acting on the beam have now been determined, and the free body diagram can be updated.

Всички външни натоварвания, действащи върху гредата, вече са определени и диаграмата на свободното тяло може да бъде актуализирана.

!!!!!!!!!Now we can start to draw the shear force and bending moment diagrams, starting from the left side of the beam. We can draw the free body diagram showing a beam segment which has been cut immediately the right of the 12 kN reaction force at Point A.

Сега можем да започнем да чертаем диаграмите на силата на срязване и момента на огъване, започвайки от лявата страна на гредата. Можем да начертаем диаграмата на свободното тяло, показваща сегмент на лъч, който е бил отрязан непосредствено отдясно на силата на реакция от 12 kN в точка А.

!!!!!!!!!!To maintain equilibrium, the shear force V(x) must balance the 12 kN reaction force:

За да се поддържа равновесие, силата на срязване V(x) трябва да балансира силата на реакция от 12 kN:

$$ V_(x) = R_A = 12 kN $$

!!!!!!!!!!Similarly, the bending moment must balance the moment generated by the 12 kN reaction load:

По същия начин огъващият момент трябва да балансира момента, генериран от 12 kN реакционно натоварване:

$$ M_(x) = R_A . x $$
$$ M_(x) = 12.2 $$
$$ M_(x) = 24 kNm $$

!!!!!!!!The shear force will be constant until we reach the next applied force, so we can draw this on shear force diagram. And the equation obtained for M(x) is the equation for a straight line, which can be drawn on the bending moment diagram.

Силата на срязване ще бъде постоянна, докато достигнем следващата приложена сила, така че можем да начертаем това на диаграмата на силата на срязване. И полученото уравнение за M(x) е уравнението за права линия, която може да бъде начертана на диаграмата на огъващия момент.

!!!!!!!!!The process is then repeated, moving the location of the imaginary cut further to the right. This time the cut is placed immediately after the 15 kN force, and the free body diagram is drawn. The shear force and bending moment can be calculated by applying the equilibrium equations.

След това процесът се повтаря, като местоположението на въображаемия разрез се премества надясно. Този път разрезът се поставя непосредствено след силата от 15 kN и се чертае диаграмата на свободното тяло. Силата на срязване и огъващият момент могат да бъдат изчислени чрез прилагане на уравненията за равновесие.

$$ M_(x) = R_A . x - F_1(x-2) $$
$$ M_(x) = 12.4 - 15(4-2) $$
$$ M_(x) = 48 - 30 $$
$$ M_(x) = 18 kNm $$

!!!!!!!!!!!The shear force and bending moment diagrams can be updated.

Диаграмите на срязващата сила и момента на огъване могат да бъдат актуализирани.

This process is repeated until the full length of the beam has been covered. The completed diagrams are shown below.

Този процес се повтаря, докато се покрие цялата дължина на гредата. Попълнените диаграми са показани по-долу.

$$ M_(x) = R_A . x - F_1(x-2) - F_2(x-4) $$
$$ M_(x) = 12.6 - 15(6-2) - 6(6-4) $$
$$ M_(x) = 72 - 60 - 12 $$
$$ M_(x) = 0 kNm $$

Checking if the sum of the forces in the vertical direction is equal to zero:

Проверка дали сборът на силите във вертикална посока е равен на нула:

$$ \sum V = 0 $$

$$ R_A - F_1 - F_2 + R_B = 0 $$
$$ 12 - 15 - 6 + 9 = 0 $$
$$ 21 -21 = 0 $$
$$ 0 = 0 $$

### Shear Force Diagram:

A shear force diagram is one which shows variation in shear force along the length of the beam. Bending moment may be defined as "the sum of moments about that section of all external forces acting to one side of that section".

Диаграмата на силата на срязване е тази, която показва изменението на силата на срязване по дължината на гредата. Моментът на огъване може да се дефинира като "сумата от моментите около тази секция на всички външни сили, действащи от едната страна на тази секция".

### Bending Moment Diagram:

The bending moment diagram shows how M (and therefore normal stress) varies across a structure. If we know the state of longitudinal or normal stress due to bending at a given section in a structure we can work out the corresponding bending moment

Диаграмата на момента на огъване показва как M (и следователно нормалното напрежение) варира в структурата. Ако знаем състоянието на надлъжно или нормално напрежение, дължащо се на огъване в дадена секция в конструкцията, можем да изчислим съответния момент на огъване.

### Normal Force Diagram:

The normal force is defined as the force that a surface employs on any other body. If that body is at rest, it means that the net force acting on the body is equal to zero. This means that the downward force, that is, the weight, must be similar to the upward force, which is the normal force.

Нормалната сила се определя като силата, която дадена повърхност прилага върху всяко друго тяло. Ако това тяло е в покой, това означава, че общата сила, действаща върху тялото, е равна на нула. Това означава, че силата надолу, тоест тежестта, трябва да бъде подобна на силата нагоре, която е нормалната сила.

The normal force is the counter force to a mass pushing against a surface. The normal force is always perpendicular to the surface. When solving problems involving normal forces, it's important to draw a free body diagram, including all of the forces acting on the mass.

Нормалната сила е противодействието на маса, която се натиска върху повърхност. Нормалната сила винаги е перпендикулярна на повърхността. Когато решавате проблеми, включващи нормални сили, е важно да начертаете диаграма на свободно тяло, включваща всички сили, действащи върху масата.

### Deflection Diagram:
Диаграма на деформациите

In structural engineering, deflection is the degree to which a part of a long structural element (such as beam) is deformed laterally (in the direction transverse to its longitudinal axis) under a load. It may be quantified in terms of an angle (angular displacement) or a distance (linear displacement). A longitudinal deformation (in the direction of the axis) is called elongation.

В структурното инженерство, деформацията е степента, до която част от дълъг конструктивен елемент (като греда) се деформира странично (в посока, напречна на надлъжната му ос) под натоварване. Може да се определи количествено по отношение на ъгъл (ъглово изместване) или разстояние (линейно изместване). Надлъжната деформация (по посока на оста) се нарича удължение.
